```python
# Initializeing the models
models = {
    'Linear Regression': LinearRegression(),
    'SVM': SVR(),
    'Random Forest': RandomForestRegressor(random_state=34),
    'XGBoost': XGBRegressor(objective='reg:squarederror', random_state=34)
}
```

### MSE measures the average of the squares of the errors between the predicted and actual values. A lower MSE means better model performance.

```python
from sklearn.metrics import mean_squared_error

for name, model in models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)
    mse = mean_squared_error(y_test, y_pred)
    print(f"{name} - MSE: {mse}")
```

### 1-RMSE is sensitive to outliers so it gives.It gives a relatively high weight to large errors because the errors are squared before they are averaged

```python
for name, model in models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    print(f"{name} - RMSE: {rmse}")
```

### Mean Absolute Error (MAE). It's the average over the test sample of the absolute differences between prediction and actual observation where all individual differences have equal weight. MAE is less sensitive to outliers than RMSE.
### it's giving smaller eerrors compeared to RMSE

```python
for name, model in models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)
    mae = mean_absolute_error(y_test, y_pred)
    print(f"{name} - MAE: {mae}")
```

### R² measures the proportion of the variance for the dependent variable that's explained by the independent variables in the model. It provides an indication of the goodness of fit of the predictions. 

```python
for name, model in models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} - R²: {r2}")
```

### Adjusted R² adjusts the statistic based on the number of independent variables in the model. It increases only if the new term improves the model more than would be expected by chance

```python
for name, model in models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)
    r2 = r2_score(y_test, y_pred)
    n = X_test_transformed.shape[0]  # Number of observations
    p = X_test_transformed.shape[1]  # Number of features
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    print(f"{name} - Adjusted R²: {adjusted_r2}")
```
